```python
from datasets import load_dataset
ds = load_dataset("coastalchp/ledgar")
```

```python
ds
```

```python
import pandas as pd

train_df = pd.DataFrame(ds['train'])
test_df = pd.DataFrame(ds['test'])
validation_df = pd.DataFrame(ds['validation'])
```

```python
X_train = train_df['text']
y_train = train_df['label']

X_test = test_df['text']
y_test = test_df['label']

X_val = validation_df['text']
y_val = validation_df['label']
```

Perform TF-IFD

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
import json
with open('class_weights.json', 'r') as f:
    class_weights = json.load(f)
```

```python
class_weights = {int(k): v for k, v in class_weights.items()}
```

Logistic Regression → simple, efficient classifier


F1 Score → handles imbalance properly


TF-IDF → converts text into meaningful numerical features

```python
# Run Baseline Model
for max_feat in [5000, 10000, 20000]:
    vectorizer = TfidfVectorizer(max_features=max_feat)
    
    X_train_tfidf = vectorizer.fit_transform(X_train)
    X_val_tfidf = vectorizer.transform(X_val)
    
    model = LogisticRegression(class_weight=class_weights, 
                               
                               max_iter=1000)
    model.fit(X_train_tfidf, y_train)
    
    preds = model.predict(X_val_tfidf)
    
    f1 = f1_score(y_val, preds, average='macro')
    
    print(f"max_features={max_feat}, Macro F1={f1}")
```

```python
#5000 features were selected for efficiency
```
